### Visualizimg Attributes Across Different Dry Bean Classes

Final Visualization with summary and attribute descriptions -  https://public.tableau.com/app/profile/chimdi.nwosu/viz/MakeoverMonday2021wk14-DryBeanAttributespython/Dashboard1

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes = True)
#Array with the colors for palette

%matplotlib inline
data = pd.read_excel('C:/Users/ffpch/Desktop/Tableau/makeover MOnday/2021/week14/makeovermonday-2021w14/makeovermonday-2021w14/Dry_Bean_Dataset.xlsx', 
                     sheet_name = 'Dry_Beans_Dataset')

import warnings
warnings.filterwarnings("ignore")
```

```python
data.columns
```

## AREA

```python
order = data.groupby('Class')['Bean ID'].count().sort_values(ascending = False)
sns.set_style("whitegrid") #set white bkgnd
#create custom palette 
colors = ["#755e4a", "#a64732","#bf7e45" , "#f2a488","#f2cda0" ,"#aaaf9f","#d3cdc8" ]
# Set custom color palette
sns.set_palette(sns.color_palette(colors))
g = sns.stripplot(data = data, x ='Class', 
                   y = 'Area', jitter = '0.35', s = 5,
                   linewidth = 0.2, edgecolor = 'black', order = order.index, alpha = 0.9)

sns.despine(left=True)#remove borders around plot

mean_width = 0.85 # set size of median  line

for bean,index in zip(g.get_xticks(), g.get_xticklabels()):
    class_name = index.get_text()
    
    mean_val = data[data['Class']==class_name].Area.median() 
    g.plot([bean-mean_width/2, bean+mean_width/2], [mean_val,mean_val], color = 'k', lw = 2 )
    
    
plt.ylabel("")
plt.xlabel("")
sns.set(font_scale = 1.3)#increase size of fonts..axes, labels
```

## COUNTS

```python
order = data.groupby('Class')['Bean ID'].count().sort_values(ascending = False).index.values
sns.set_style("whitegrid") #set white bkgnd
#create custom palette 
colors = ["#755e4a", "#a64732","#bf7e45" , "#f2a488","#f2cda0" ,"#aaaf9f","#d3cdc8" ]
# Set custom color palette
sns.set_palette(sns.color_palette(colors))
ax = sns.countplot(x="Class", data=data,order = order, dodge = False, linewidth = 0.5,);    
sns.despine(left=True)#remove borders around plot
    
for i in ax.patches:
    height = i.get_height()#get height
    ax.text(x=i.get_x()+(i.get_width()/2),#add text to each bar, get x position + 1/2 way through bar to end on center
    y = height+50,# y position with elevation of 50 for label
    s='{:.0f}'.format(height), ha = 'center');#ignores decimal and aligns to center
    
plt.ylabel("Number of Beans");
plt.xlabel("Bean Class");
sns.set(font_scale = 1.3)#increase size of fonts..axes, labels
plt.savefig('C:/Users/ffpch/Desktop/Tableau/makeover MOnday/2021/week14/makeovermonday-2021w14/makeovermonday-2021w14/figure7.png')
```
